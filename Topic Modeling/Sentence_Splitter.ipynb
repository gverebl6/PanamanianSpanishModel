```
import re
```

```
txt=open('data/2015_03_18_A_COMI_AGRO.txt')
```

```
txt.seek(0)
txtText=txt.read()
finalSentences=[]
```

```
p=re.compile('[A-z]+[^.!?]*[.!?]')
sentences=p.findall(txtText)
```

```
for sentence in sentences:
  if len(sentence.split()) > 1: #Eliminar oraciones de una sola palabra.
    sentenceFixed=sentence.replace('\n','')
    finalSentences.append(sentenceFixed)
```

```
print(finalSentences)
```
